```python
df = pd.read_excel("mega_data_cleaning_dataset.xlsx")
print(df.head())
```

```python
df.shape
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.drop_duplicates(subset="Customer_ID", inplace=True)
df.shape
```

```python
df.drop_duplicates(subset="Email", inplace=True)
df.shape
```

```python
df.drop_duplicates(subset="Product_ID", inplace=True)
df.shape
```

```python
df["Name"].fillna("Unknown", inplace=True)
```

```python
df["Email"].fillna("noemail@example.com", inplace=True)
```

```python
df["Product_Name"].fillna("Misc", inplace=True)
```

```python
df["Category"].fillna("Misc", inplace=True)
```

```python
df["Feedback_Score"].fillna(df.groupby("Product_ID")["Feedback_Score"].transform("median"), inplace=True)
```

```python
df["Quantity"].fillna(1, inplace=True)
```

```python
df = df[df["Price"]>0]
```

```python
df = df[df["Quantity"]>0]
```

```python
df = df[(df['Age'] >= 0) & (df['Age'] <= 120)]
```

```python
df = df[df['Delivery_Date'] >= df['Purchase_Date']]
```

```python
df['Gender'] = df['Gender'].replace({'Mle':'Male','m':'Male','f':'Female','F':'Female'})
```

```python
df['Payment_Type'] = df['Payment_Type'].replace({'crdit':'Credit Card','deb':'Debit Card'})
```

```python
df['Account_Status'] = df['Account_Status'].str.capitalize().replace({'Act':'Active','Inact':'Inactive'})
```

```python
df['Total_Purchase'] = df['Price'] * df['Quantity']
```

```python
df['Discounted_Price'] = df['Price'] - df['Discount']
```

```python
df['Delivery_Duration'] = (df['Delivery_Date'] - df['Purchase_Date']).dt.days
```

```python
bins = [0, 12, 19, 59, 120]
labels = ['Child','Teen','Adult','Senior']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
df['Loyalty_Level'] = pd.qcut(df['Total_Purchase'], q=3, labels=['Low','Medium','High'])
```

```python
df['Gender_Code'] = df['Gender'].map({'Male':1,'Female':2}).fillna(3)
```

```python
df['Account_Code'] = df['Account_Status'].map({'Active':1,'Inactive':2,'Banned':3})
```

```python
df['Country'] = df['Country'].str.title()
```

```python
df['Payment_Type'] = df['Payment_Type'].str.title()
```

```python
df = df.applymap(lambda x: x.strip().lower() if isinstance(x, str) else x)
```
